```python
import pandas as pd
```

```python
# import clean keywords csv
url = 'https://raw.githubusercontent.com/Megreid23/final_project/main/csv%20files/keywords.csv'
keywords = pd.read_csv(url)
keywords
```

```python
keywords.drop(columns = ['Unnamed: 0'], inplace=True)
keywords.head()
```

```python
# convert list-like object columns into lists
# https://towardsdatascience.com/dealing-with-list-values-in-pandas-dataframes-a177e534f173
keywords['keywords'] = keywords['keywords'].apply(eval)
keywords
```

```python
# separate list-like column into separate columns
# https://stackoverflow.com/questions/50264678/explode-column-of-lists-into-multiple-columns
series = keywords['keywords'].apply(pd.Series)
series
```

```python
# merge keywords table and series
# https://stackoverflow.com/questions/26265819/how-to-merge-a-series-and-dataframe/40762674#40762674
keywords_merged = keywords.merge(series, left_index=True, right_index=True)
keywords_merged.head()
```

```python
# drop original list column
keywords_merged.drop(columns = ['keywords'], inplace=True)
keywords_merged.head()
```
