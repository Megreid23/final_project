```python
import pandas as pd
```

```python
# import clean studios csv
url = 'https://raw.githubusercontent.com/Megreid23/final_project/main/csv%20files/studios.csv'
studios = pd.read_csv(url)
studios.head()
```

```python
studios.drop(columns = ['Unnamed: 0'], inplace=True)
studios.head()
```

```python
studios.info()
```

```python
# convert list-like object columns into lists
# https://towardsdatascience.com/dealing-with-list-values-in-pandas-dataframes-a177e534f173
studios['studios'] = studios['studios'].apply(eval)
studios.head()
```

```python
# separate list-like column into separate columns
# https://stackoverflow.com/questions/50264678/explode-column-of-lists-into-multiple-columns
series = studios['studios'].apply(pd.Series)
series.head()
```

```python
# merge studios table and series
# https://stackoverflow.com/questions/26265819/how-to-merge-a-series-and-dataframe/40762674#40762674
studios_merged = studios.merge(series, left_index=True, right_index=True)
studios_merged
```

```python
studios_merged.head()
```

```python
# drop original list column
studios_merged.drop(columns = ['studios'], inplace=True)
studios_merged.head()
```

```python
studios_merged.info()
```
